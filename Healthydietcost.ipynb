```python
import pandas as pd
df = pd.read_csv('price_of_healthy_diet_clean.csv')
df
```

```python
df.dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.scatterplot(x = 'cost_vegetables_ppp_usd', y = 'cost_healthy_diet_ppp_usd', data = df)
plt.xlabel('Cost of veggies')
plt.ylabel('Cost of Healthy Diet (PPP USD)')
plt.title('Cost of Healthy Diet wrt cost of veg')
plt.show()
sns.scatterplot(x = 'cost_fruits_ppp_usd', y = 'cost_healthy_diet_ppp_usd', data = df)
plt.xlabel('cost of fruits')
plt.ylabel('Cost of Healthy Diet (PPP USD)')
plt.title('Cost of Healthy Diet wrt cost of fruits')
plt.show()
```
